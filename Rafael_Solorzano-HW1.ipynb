## Excercise 1: Transcription

Create a function called `dna2rna` that takes the `coronaGenome` and converts all of the `t` into `u`. This function is converting DNA into RNA and more accurately represents the genome of the corona virus since it has an RNA-based genome. Save the output of the function into a new variable called `coronaRNAgenome`.



```python
#For loop 
def dna2rna(coronaGenome): #creating function that takes in the coronaGenome 
    return(coronaGenome.replace('t','u')) # replacing all instances of 't' with 'u' within coronaGenome to make it RNA

coronaRNAgenome = dna2rna(coronaGenome) # storing the "transcribed" genome into a newly named variable we can use later 
coronaRNAgenome 
```

## Excercise 3: Mutation

Now let's simulate the process of mutation. This is a random process so we are going to have to bring in python's random number generator.

Next create a function called `mutate` that iterates through every base of the `coronaRNAgenome` and depending on a second parameter called `mutationRate` uses the `rd.choice` function to randomly change a base pair. HIV has a mutation rate of 10^-3 mutations per base pair per replication cycle whearas covid-19 is has been estimated to be closer to 10^-4 to 10^-5 mut/bp/cycle. 

HINT: One possible implimentation could use the following line:
`if rd.random() > mutationRate:`
This line will most likely be paired with an `else` statement.

```python

import random as rd  #we are importing the random function 


def mutate(genome):  #creating new function that "induces" random mutations at a specified rate 
    mutationRate = 1e-4 #We are defining the new mutation rate 
    mutations = "" #storage for the new mutated genome 
    for i in genome: # for loop: for an index position 'i' we... 
        rd.random()  # generate a random number between 0-1
        if rd.random() < mutationRate: #if our randomly generated number is lower than the mutation rate, then we choose to "induce" a mutation 
            n = rd.choice('AUGC') # If it's deemed that we need to induce, then we randomly choose from AUGC
            mutations += n # We append the mutation
        else: #if the we do not induce a mutation...
            mutations += i  #then append the next nucleotide that should have come in the sequence 
    return mutations # output the mutated genome
            
print(mutate(coronaRNAgenome))
    
```

## Excercise 4: Translation and the Spike

Finally, in order for the coronavirus to manipulate the cells in our bodies parts of its RNA genome must be translated into protein.

Build a function called `RNA2Protein()` that translates any sequence of RNA to protein using the codon table.


```python
def RNA2Protein(seq):
    
    RNACodonTable = {
        "UUU":"F", "UUC":"F", "UUA":"L", "UUG":"L",
    "UCU":"S", "UCC":"s", "UCA":"S", "UCG":"S",
    "UAU":"Y", "UAC":"Y", "UAA":"STOP", "UAG":"STOP",
    "UGU":"C", "UGC":"C", "UGA":"STOP", "UGG":"W",
    "CUU":"L", "CUC":"L", "CUA":"L", "CUG":"L",
    "CCU":"P", "CCC":"P", "CCA":"P", "CCG":"P",
    "CAU":"H", "CAC":"H", "CAA":"Q", "CAG":"Q",
    "CGU":"R", "CGC":"R", "CGA":"R", "CGG":"R",
    "AUU":"I", "AUC":"I", "AUA":"I", "AUG":"M",
    "ACU":"T", "ACC":"T", "ACA":"T", "ACG":"T",
    "AAU":"N", "AAC":"N", "AAA":"K", "AAG":"K",
    "AGU":"S", "AGC":"S", "AGA":"R", "AGG":"R",
    "GUU":"V", "GUC":"V", "GUA":"V", "GUG":"V",
    "GCU":"A", "GCC":"A", "GCA":"A", "GCG":"A",
    "GAU":"D", "GAC":"D", "GAA":"E", "GAG":"E",
    "GGU":"G", "GGC":"G", "GGA":"G", "GGG":"G",
    }
    
    
    
    protein = "" #created an empty string for the protein     

    if len(seq)%3 == 0: #if the length of the RNA seq is a multiple of 3, then we can translate 
        for i in range(0, len(seq), 3): #this set's the entire range of input to be the length of the genome, and to index in sets of 3, a codon 
            codon = seq[i: i + 3] #this sets a codon to be a set of 3 nucleotides 
            protein += RNACodonTable[codon] #building the protein: based on codon table, we add to protein string
    else: 
        raise Exception('Sequence is not a multiple of 3!!!') #if the genome sequence is not a multiple of 3 we have this warning
    
    return protein 
    
    
RNA2Protein('AUGCAAUUG')
```
