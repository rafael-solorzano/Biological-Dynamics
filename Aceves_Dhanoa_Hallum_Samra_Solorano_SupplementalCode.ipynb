# Asymptotic Stability of a Modified Lotka-Volterra Model with Small Immigration and Catastrophe

```python
# Gather all necessary libraries for manipulation of data, integration, and plotting
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

## Assigning Variables
 
* x - prey population
* y - predator population
* r - reproduction rate of prey
* b - birth rate of predator per prey captured
* a - rate at which predators kill prey
* m - predator mortality rate
* C - small immigration into prey population constant 
* D - small immigration into predator population constant
* h - functional responce coefficient 

For our cases, we define the C(x) and D(y) as specified in the paper.

### Linear

1) Prey Immigrants, C(x) = C and D(y) = 0

2) Predator Immigrants, C(x) = 0 and D(y) = d 

3) Few Prey Immigrants, C(x) = c/x and D(y) = 0

4) Few Predator Immigrants, C(x) = 0 and D(y) = c/x 

5) No immigration or emigration of prey/predators C(x) = 0 and D(y) = 0

### Hyperbolic

1) Prey Immigrants, C(x) = C and D(y) = 0

2) Predator Immigrants, C(x) = 0 and D(y) = d 

3) Few Prey Immigrants, C(x) = c/x and D(y) = 0

4) Few Predator Immigrants, C(x) = 0 and D(y) = c/x 

5) No immigration or emigration of prey/predators C(x) = 0 and D(y) = 0

### Sigmoid Functional 

1) Prey Immigrants, C(x) = C and D(y) = 0

2) Predator Immigrants, C(x) = 0 and D(y) = d 

3) Few Prey Immigrants, C(x) = c/x and D(y) = 0

4) Few Predator Immigrants, C(x) = 0 and D(y) = c/x 

5) No immigration or emigration of prey/predators C(x) = 0 and D(y) = 0



## Linear

```python
#  Linear No Migration; C(x) = 0 & D(x) = 0 
def LV_model (state, t0, r = 0.1, a= 0.1, b = 0.3 ,m= 0.2, C=0, D =0):
    x = state[0]
    y = state[1]
    
    
    
    dxdt = (r*x) - (a*x*y) + C
    
  
    dydt = (b*x*y) - (m*y) +D
    

    
    return(dxdt,dydt)

t = np.linspace(0,2000,10000)
state0 = np.array([5,5])

state = odeint(LV_model, state0 ,t)
plt.plot(t,state)
plt.legend(('Prey Population', 'Predator Population'))
plt.ylabel('Population (Individuals)')
plt.xlabel('Time (Years)')
plt.title('Linear Lotka-Volterra Model with No Immigration')
```

# Adding Catastrophe to Type I, II, and III LV Models with immigration of predators/prey. 

```python
# Catastrophe simulation of Type I, II, and III LV Models with Prey Immigration (Brina)

      # Catastrophe Simulation of Type I Model with Prey Immigrants 

state0 = np.array([7,2])  #initial conditions
def LV_model (state, t0, r = 0.1, a= 0.1, b = 0.3 ,m= 0.2, C=(0.01), D =0):
    x = state[0]
    y = state[1]
    
    dxdt = (r*x) - (a*x*y) + C   #small immigration C > 0
    dydt = (b*x*y) - (m*y) +D    #no immigration D = 0
    
    return(dxdt,dydt)


t1 = np.linspace(0,1000,5000)       #create timespace prior to catastrophe

state = odeint(LV_model, state0, t1)    #integrate over initial timespace


state1 = 0.5*state[1000]        #grab x and y (prey, predator values) @ timepoint 1000 and reduce both by 50% due to natural disaster/catastrophe

def LV_postcat (state1, t0, r = 0.1, a= 0.1, b = 0.3 ,m= 0.2, C=(0.01), D =(0)):
    x1 = state1[0]
    y1 = state1[1]
    
    dxdt1 = (r*x1) - (a*x1*y1) + C      #same differential equations with new initial conditions (50% of populations die at t = 1000 )
    dydt1 = (b*x1*y1) - (m*y1) + D
    
    return(dxdt1,dydt1)
    
t2 = np.linspace(1000,2000,5000)        #create timespace post-catastrophe
state_post = odeint(LV_postcat, state1, t2)      #integrate post catastrophe

plt.plot(t1,state)
plt.plot(t2,state_post)    #plot graphs of populations both before and after catastrophe on same figure


plt.legend(('Prey Population', 'Predator Population', 'Prey Population s/p Catastrophe', 'Predator Population s/p Catastrophe'))
plt.ylabel('Population (Individuals)')
plt.xlabel('Time (Years)')
plt.title('Linear Lotka-Volterra Model with Prey Immigration with a Catastrophe at 1000 years')
plt.show()


      # Catastrophe Simulation of Type II Model with Prey Immigrants 

state0 = np.array([7,2])
def LV_model (state, t0, r = 0.1, a= 0.1, b = 0.3 ,m= 0.2, C=(0.01), D =0, h = 0.01):
    x = state[0]
    y = state[1]
    
    dxdt = (r*x) - ((a*x*y) / (1+h*x)) + C     
    
    dydt = ((b*x*y) / (1+h*x)) - (m*y) + D  

    return(dxdt,dydt)


t1 = np.linspace(0,1000,5000)      

state = odeint(LV_model, state0, t1)   


state1 = 0.5*state[1000]   

def LV_postcat (state1, t0, r = 0.1, a= 0.1, b = 0.3 ,m= 0.2, C=(0.01), D =0, h = 0.01):
    x1 = state1[0]
    y1 = state1[1]
    
    dxdt1 = (r*x1) - ((a*x1*y1) / (1+h*x1)) + C   
    
    dydt1 = ((b*x1*y1) / (1+h*x1)) - (m*y1) + D
    
    return(dxdt1,dydt1)
    
t2 = np.linspace(1000,2000,5000)    
state_post = odeint(LV_postcat, state1, t2)  

plt.plot(t1,state)      
plt.plot(t2,state_post)       

plt.legend(('Prey Population', 'Predator Population', 'Prey Population s/p Catastrophe', 'Predator Population s/p Catastrophe'))
plt.ylabel('Population (Individuals)')
plt.xlabel('Time (Years)')
plt.title('Hyperbolic Lotka-Volterra Model with Prey Immigration with a Catastrophe at 1000 years')
plt.show()

      # Catastrophe Simulation of Type III Model with Prey Immigrants 

state0 = np.array([7,2])
def LV_model_Sigmoidal (state, t0, r = 0.1, a= 0.1, b = 0.3 ,m= 0.2, C=(0.01), D =0, h = 0.01):
    x = state[0]
    y = state[1]
    
    dxdt = (r*x) - ((a*(x**2)*y) / (1+h*(x**2))) + C
    
    dydt = ((b*(x**2)*y) / (1+h*(x**2))) - (m*y) + D
    
    return(dxdt,dydt)


t1 = np.linspace(0,1000,5000)

state = odeint(LV_model_Sigmoidal, state0, t1)


state1 = 0.5*state[1000]

def LV_postcat_Sig (state1, t0, r = 0.1, a= 0.1, b = 0.3 ,m= 0.2, C=(0.01), D =0, h = 0.01):
    x1 = state1[0]
    y1 = state1[1]
    
    dxdt1 = (r*x1) - ((a*(x1**2)*y1) / (1+h*(x1**2))) + C
    
    dydt1 = ((b*(x1**2)*y1) / (1+h*(x1**2))) - (m*y1) + D
    
    return(dxdt1,dydt1)
    
t2 = np.linspace(1000,2000,5000)
state_post = odeint(LV_postcat_Sig, state1, t2)

plt.plot(t1,state)
plt.plot(t2,state_post)

plt.legend(('Prey Population', 'Predator Population', 'Prey Population s/p Catastrophe', 'Predator Population s/p Catastrophe'))
plt.ylabel('Population (Individuals)')
plt.xlabel('Time (Years)')
plt.title('Sigmoidal Lotka-Volterra Model with Prey Immigration with a Catastrophe at 1000 years')
plt.show()
```
